```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Cliente":   [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":      [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":  [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras":   [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})

# 1) Análisis Exploratorio
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

sns.boxplot(x=df["Compras"])
plt.title("Boxplot de Compras")
plt.show()

sns.scatterplot(x=df["Ingresos"], y=df["Compras"])
plt.title("Relación entre Ingresos y Compras")
plt.show()

# 2) Visualización Avanzada
sns.pairplot(df[["Edad", "Ingresos", "Compras"]])
plt.show()

# Mapa de calor de correlaciones
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Variables")
plt.show()

# Violin Plot de Compras por Categoría
sns.violinplot(x="Categoría", y="Compras", data=df, hue="Categoría", dodge=False, palette="muted", legend=False)
plt.title("Distribución de Compras por Categoría")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Simulamos datos para horas trabajadas y productividad
np.random.seed(42)
data = {
    'Horas_Trabajadas': np.random.uniform(4, 12, 50),
    'Productividad': np.random.normal(70, 15, 50)
}
df_prod = pd.DataFrame(data)

sns.set_theme(style="darkgrid") 
plt.figure(figsize=(8, 5))


# Scatter plot con regresión para ver la relación (Elegi regresion porque permite identificar la direccionalidad y la fuerza de la relación entre las variables)
sns.regplot(data=df_prod, x='Horas_Trabajadas', y='Productividad', 
    scatter_kws={'alpha':0.6}, line_kws={'color':'red'})

plt.title("Relación: Horas Trabajadas vs Productividad")
plt.show()
```

#### Ejercicio para la Sección 3
Objetivo: Aplicar gráficos avanzados para extraer insights de un dataset.

Instrucciones:

1. Carga el dataset de diamantes de Seaborn (diamonds).
2. Crea un Pairplot para analizar la relación entre precio, quilates y profundidad.
3. Genera un Heatmap para visualizar la correlación entre las variables numéricas.
4. Crea un Violin Plot para comparar la distribución del precio en cada categoría de corte (cut).
5. Escribe una conclusión basada en los gráficos.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar dataset
df = sns.load_dataset("diamonds")

# 1. Pairplot entre precio, quilates y profundidad
sns.pairplot(df[["price", "carat", "depth"]])
plt.show()

# 2. Heatmap de correlaciones
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Diamantes")
plt.show()

# 3. Violin Plot para comparar precio según corte
sns.violinplot(x="cut", y="price", data=df, hue="cut", dodge=False, palette="muted", legend=False)
plt.title("Distribución del Precio por Tipo de Corte")
plt.show()
```

#### Conclusión:
La principal conclusión es que el precio del diamante está determinado casi exclusivamente por sus dimensiones físicas y su peso (carat), dejando a las variables de calidad (como el corte) en un segundo plano estadístico.

1. Análisis Numérico
- Dominio del Quilate: Existe una correlación positiva muy fuerte (0.92) entre el carat y el price. Es el motor principal del valor.
- Redundancia Dimensional: Las variables x, y y z tienen correlaciones superiores al 0.86 con el precio y casi perfectas entre sí.
- Variables Irrelevantes: La profundidad (depth) y la tabla (table) no tienen impacto lineal en el precio (-0.01 y 0.13 respectivamente)

2. Análisis Categórico
- El gráfico de violín revela un fenómeno contraintuitivo: los diamantes con corte "Fair" (pobre) presentan medianas de precio más altas que los de corte "Ideal". Esto no significa que el mal corte sea más caro, sino que existe una variable confusora. Los diamantes con cortes de menor calidad suelen ser piedras mucho más grandes donde se priorizó mantener el peso (carat) sobre la perfección del tallado.
- Distribución: El corte "Ideal" tiene una base muy ancha en precios bajos, lo que indica una sobreoferta de diamantes pequeños de alta calidad.
