```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear datos ficticios
df = pd.DataFrame({
    "Edad":      [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos":  [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000],
    "Categoría": ["Joven", "Adulto", "Adulto", "Adulto", "Senior", "Senior", "Joven", "Adulto", "Adulto", "Senior"]
})

# Histograma de edades
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Boxplot de ingresos
sns.boxplot(x=df["Ingresos"])
plt.title("Boxplot de Ingresos")
plt.show()

# Scatter Plot entre Edad e Ingresos
sns.scatterplot(x=df["Edad"], y=df["Ingresos"], hue=df["Categoría"])
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

## Conclusión

- El histograma nos permite ver la distribución de edades.
- El boxplot ayuda a detectar posibles valores atípicos en ingresos.
- El scatter plot muestra la relación entre Edad e Ingresos, diferenciando categorías.

####  2: Matplotlib y Seaborn
2.1. ¿Qué es Matplotlib y cómo usarlo?
Matplotlib es la biblioteca más básica y flexible para crear gráficos en Python. Se usa principalmente para gráficos simples como líneas, barras y dispersión.

```python
#Ejemplo en Python – Crear un gráfico de líneas con Matplotlib
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

# Crear gráfico de líneas
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Crecimiento")
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.title("Ejemplo de Gráfico de Líneas")
plt.legend()
plt.show()
```

### 2.2. ¿Qué es Seaborn y cómo usarlo?
Seaborn está basado en Matplotlib, pero ofrece una sintaxis más sencilla y gráficos más atractivos por defecto. Se usa para visualizaciones estadísticas avanzadas, como distribuciones de datos, correlaciones y visualización de múltiples variables.

Ejemplo en Python – Histogramas y Boxplots con Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame con datos ficticios
df = pd.DataFrame({
    "Edad":     [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

# Histograma de edades
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Boxplot de ingresos
sns.boxplot(x=df["Ingresos"])
plt.title("Boxplot de Ingresos")
plt.show()
```

### 📊 Comparación: Matplotlib vs. Seaborn

| Característica | Matplotlib | Seaborn |
| :--- | :--- | :--- |
| **Facilidad de uso** | Más flexible pero requiere más código. | Más intuitivo y automático. |
| **Personalización** | Muy alta (permite modificar cada detalle). | Alta, pero con menos código necesario. |
| **Aplicaciones** | Gráficos básicos y personalizables. | Gráficos estadísticos más avanzados. |
| **Ejemplo típico** | Gráficos de líneas y barras. | Histogramas, heatmaps y pairplots. |

¿Cuándo usar cada uno?

Usa Matplotlib cuando necesites total control sobre los detalles del gráfico.
Usa Seaborn cuando necesites gráficos estadísticos listos para usar con menos código.

#### Pairplot: Exploración Multivariable
¿Qué es un Pairplot?
Es una herramienta que muestra relaciones entre múltiples variables en un dataset de manera eficiente.

¿Cuándo usarlo?

Cuando queremos ver cómo se relacionan múltiples variables numéricas en un solo gráfico.
Cuando queremos detectar correlaciones visualmente sin calcular matrices de correlación.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar dataset de pingüinos de Seaborn
df = sns.load_dataset("penguins")

# Crear Pairplot
sns.pairplot(df, hue="species", diag_kind="kde")
plt.show()
```

##### Conclusión:

Cada gráfico en la diagonal muestra la distribución univariada de cada variable.
Los gráficos fuera de la diagonal muestran relaciones entre pares de variables.

#### Heatmaps: Mapas de Correlación
¿Qué es un Heatmap?
- Es un mapa de calor que muestra cómo se correlacionan las variables numéricas en un dataset.

¿Cuándo usarlo?

- Cuando queremos analizar fuerza y dirección de correlación entre variables.
- Cuando buscamos variables redundantes que pueden eliminarse o agruparse.

```python
#Ejemplo en Python – Heatmap de correlaciones

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar dataset
df = sns.load_dataset("penguins").dropna()

# Calcular matriz de correlación
correlaciones = df.corr(numeric_only=True)

# Crear Heatmap
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación entre Variables")
plt.show()
```

Conclusión:

- Un coeficiente cercano a 1 indica alta correlación positiva (ejemplo: peso y tamaño).
- Un coeficiente cercano a -1 indica alta correlación negativa (ejemplo: velocidad y tiempo de viaje).
- Un coeficiente cercano a 0 significa que las variables no están relacionadas.

#### Violin Plots: Comparación de Distribuciones
¿Qué es un Violin Plot?
- Es una mezcla entre un boxplot y un histograma, que muestra la distribución de los datos dentro de cada categoría.

¿Cuándo usarlo?

- Cuando queremos ver cómo varían los valores en diferentes grupos.
- Cuando queremos detectar asimetrías en la distribución de datos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset
df = sns.load_dataset("penguins")

# Crear Violin Plot
sns.violinplot(x="species", y="bill_length_mm", data=df, hue="species", palette="muted", legend=False)
plt.title("Distribución del Largo del Pico por Especie")
plt.show()
```

#### Conclusión:

El ancho del violín representa dónde se encuentran más datos.
Las colas indican valores extremos o atípicos.
