## 2. Tipos de Análisis Exploratorio de Datos

Existen dos tipos principales de análisis exploratorio:

| Tipo de Análisis | Descripción | Ejemplo |
| :--- | :--- | :--- |
| **Análisis Univariado** | Examina una variable a la vez para entender su distribución. | Histograma de edades de clientes. |
| **Análisis Multivariado** | Analiza relaciones entre múltiples variables. | Correlación entre edad e ingresos en un dataset de clientes. |

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos de ejemplo
df = pd.DataFrame({
    "Edad":     [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

# Análisis Univariado: Histograma de edades
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Análisis Multivariado: Diagrama de dispersión entre Edad e Ingresos
sns.scatterplot(x=df["Edad"], y=df["Ingresos"])
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

## Análisis Univariado

El análisis univariado examina una sola variable a la vez, permitiendo entender su distribución, tendencia y presencia de valores atípicos.

### Técnicas Comunes

| Técnica | Descripción | Ejemplo |
| :--- | :--- | :--- |
| **Medidas de tendencia central** | Media, mediana y moda para analizar la distribución de la variable. | ¿Cuál es el salario promedio de los empleados? |
| **Medidas de dispersión** | Rango, varianza y desviación estándar para analizar la variabilidad de los datos. | ¿Cuánto varían las edades de los clientes? |
| **Histogramas y Boxplots** | Visualización de distribuciones y detección de outliers. | ¿Los ingresos tienen una distribución normal? |

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame con datos de ejemplo
df = pd.DataFrame({
    "Edad":     [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

# Medidas estadísticas básicas
print(df["Edad"].describe())

# Histograma para analizar la distribución
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Boxplot para detectar outliers
sns.boxplot(x=df["Edad"])
plt.title("Boxplot de Edades")
plt.show()
```

##  Ejercicio para la Sección 2

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Empleado":    [1, 2, 3, 4, 5],
    "Edad":        [25, 40, 35, 50, 30],
    "Salario":     [50000, 60000, 70000, 80000, 45000],
    "Experiencia": [2, 10, 7, 15, 3]
})

# Análisis Univariado: Estadísticas básicas
print(df["Salario"].describe())

# Histograma de salarios
sns.histplot(df["Salario"], bins=5, kde=True)
plt.title("Distribución de Salarios")
plt.show()

# Boxplot para detectar outliers en Salario
sns.boxplot(x=df["Salario"])
plt.title("Boxplot de Salarios")
plt.show()

# Análisis Multivariado: Correlación entre Edad y Salario
print(df[["Edad", "Salario"]].corr())

# Gráfico de dispersión
sns.scatterplot(x=df["Edad"], y=df["Salario"])
plt.title("Relación entre Edad y Salario")
plt.show()
```
