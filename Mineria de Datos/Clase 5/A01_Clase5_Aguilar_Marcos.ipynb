### Ejercicio Integrador y Publicación en el Foro

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder

# Crear DataFrame
df = pd.DataFrame({
    "Cliente":          [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":             [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":         [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Historial_Pagos":  ["Bueno", "Regular", "Malo", "Bueno", "Regular", "Malo", "Bueno", "Regular"],
    "Categoría":        ["A", "B", "C", "A", "B", "C", "A", "B"]
})

# 1) Manejo de valores nulos
df.fillna(df.median(numeric_only=True), inplace=True)

# 2) Detección de outliers con IQR
Q1 = df["Ingresos"].quantile(0.25)
Q3 = df["Ingresos"].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[(df["Ingresos"] >= Q1 - 1.5*IQR) & (df["Ingresos"] <= Q3 + 1.5*IQR)].copy()

# 3) Transformación numérica
scaler_minmax = MinMaxScaler()
df_cleaned["Ingresos_Normalizado"] = scaler_minmax.fit_transform(df_cleaned[["Ingresos"]])

scaler_std = StandardScaler()
df_cleaned["Edad_Estandarizada"] = scaler_std.fit_transform(df_cleaned[["Edad"]])

# 4) Transformación categórica
df_cleaned = pd.get_dummies(df_cleaned, columns=["Historial_Pagos"], prefix="Historial")

le = LabelEncoder()
df_cleaned["Categoría_Label"] = le.fit_transform(df_cleaned["Categoría"])

print("Datos finales transformados:")
display(df_cleaned.head())
```

### 3) Interacción con una Herramienta de IA (LLM)

**Consulta realizada:**
- "¿Qué métodos avanzados de imputación de datos recomiendas para conjuntos de datos con muchas variables numéricas y categóricas?"

**Respuesta de la IA (Resumen):**
Para datasets complejos, la IA recomienda técnicas más allá de la media o mediana:
- **MICE (Multivariate Imputation by Chained Equations):** Modela cada variable con nulos como una función de las demás.
- **KNN Imputer:** Encuentra los "vecinos más cercanos" para estimar el valor faltante basándose en similitudes.
- **Iterative Imputer:** Una versión de Scikit-learn inspirada en MICE que usa regresión para predecir nulos.

**Reflexión y Comparación:**
En clase aprendimos métodos básicos como la eliminación de filas o la imputación por medidas de tendencia central (media/mediana). La respuesta de la IA es sumamente útil porque introduce el concepto de **imputación predictiva**. 

Sin embargo, como estudiante de **Data Science**, entiendo que aplicar MICE o KNN requiere mucho más poder de cómputo y puede introducir ruido si no se ajusta bien. Para el dataset pequeño de este ejercicio, la mediana fue la opción más eficiente, pero para proyectos de **Prometech** con grandes volúmenes de datos, considerar técnicas multivariadas sería el siguiente paso lógico.
