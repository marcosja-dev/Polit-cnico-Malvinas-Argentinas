```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib widget

x = np.linspace(0, 2, 100)

plt.plot(x, x, 'pink', label='lineal')
plt.plot(x, x**2, label='cuadratica')
plt.plot(x, x**3, label='cubica')
plt.legend()

#Descripcion de títulos y ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title("Estructura de gráfico")

#Límites en ejes
plt.xlim(0,2)
plt.ylim(0,8)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Creamos una figura más grande para que no se amontonen
plt.figure(figsize=(10, 6))

# Filtramos marcadores que tienen una representación visual real
valid_markers = {k: v for k, v in Line2D.markers.items() if v != 'nothing' and v != 'pixel'}

for i, (marker, description) in enumerate(valid_markers.items()):
    # i % 10 nos da la columna, i // 10 nos daría una grilla más prolija
    plt.scatter(i % 10, i // 10, marker=marker, s=200, label=str(marker))
    
    # Agregamos el símbolo como texto al lado para saber cuál es cuál
    plt.text((i % 10) + 0.2, i // 10, str(marker), fontsize=9)

plt.title("Zoológico de Marcadores en Matplotlib")
plt.axis('off') # Quitamos los ejes para que parezca un catálogo
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
datos_mb,features = make_blobs(n_samples=20, centers=4, random_state=3)
plt.scatter(datos_mb[:, 0], datos_mb[:, 1], marker='o', c=features)
plt.show()
```

### Parámetros de la función `make_blobs`

| Parámetro | Descripción | Valor por defecto |
| :--- | :--- | :--- |
| **n_samples** | El número total de puntos divididos en partes iguales entre los grupos. | 100 |
| **centers** | El número de centros a generar o las ubicaciones de los centros fijos. | 3 |
| **cluster_std** | La desviación estándar de los grupos (dispersión). | 1.0 |
| **shuffle** | Mezclar las muestras aleatoriamente. | True |
| **random_state** | Semilla para el generador de números aleatorios (asegura reproducibilidad). | None |
