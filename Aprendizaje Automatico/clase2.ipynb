```python
#2. Cargar los datos de clientes desde el archivo JSON y mostrar la cantidad total de clientes.
import pandas as pd

clientes = pd.read_json('archivos_semana_2/clientes.json')
#print(clientes)

totalClientes = len(clientes)
print(f"La cantidad total de clientes es {totalClientes}")

```

```python
#3. Cargar los datos de inventario desde la hoja de cálculo Excel y calcular el promedio de stock disponible para todos los productos.
import pandas as pd

inventario = pd.read_excel('archivos_semana_2/inventario.xlsx')
#print(inventario)

promedio_stock = inventario['stock'].mean()
print(f"El promedio del stock disponble es: {promedio_stock}")
```
