```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 1. Generación de un Dataset Sintético Realista (500 clientes)
np.random.seed(42)
n_samples = 500

edad = np.random.randint(18, 65, n_samples)
ingresos = np.random.randint(1500, 9500, n_samples)
tickets = np.random.randint(0, 10, n_samples)

# Creamos una lógica para el Churn (más tickets y menos ingresos -> más probabilidad de irse)
# Le sumamos ruido normal para que no sea una regla matemática perfecta y los modelos tengan que aprender
score = (tickets * 1.2) - (ingresos / 2500) + (edad / 45) + np.random.normal(0, 1.5, n_samples)
churn = (score > 1.5).astype(int)

data = pd.DataFrame({
    "Edad": edad,
    "Ingresos_Mensuales": ingresos,
    "Tickets_Soporte": tickets,
    "Churn": churn
})

# Separar características (X) y etiqueta (y)
X = data[["Edad", "Ingresos_Mensuales", "Tickets_Soporte"]]
y = data["Churn"]

# 2. Partición de datos (70/30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Escalado de características (Estandarización)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Entrenar y evaluar K-NN (k=5)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)

# 5. Entrenar y evaluar Árbol de Decisión
tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

# 6. Salida de métricas para tu entrega
print("=== METRICAS K-NN ===")
print(f"Accuracy: {accuracy_score(y_test, y_pred_knn):.4f}")
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_knn))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_knn))

print("\n=== METRICAS ARBOL DE DECISION ===")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tree):.4f}")
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_tree))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_tree))
```
