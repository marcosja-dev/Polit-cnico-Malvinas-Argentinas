# 2. Gráfico de Dispersión

```python
import pandas as pd
import matplotlib.pyplot as plt

autosdf = pd.read_csv('archivos_semana_3/Automobile.csv')
autosdf = autosdf.dropna()
autosdf.plot.scatter(x='weight', y='mpg', alpha=0.5, color='blue', figsize=(10, 6))

plt.title='Relación entre Peso del Automóvil y Consumo (MPG)',
plt.xlabel='Peso del Automóvil',
plt.ylabel='Millas por Galón (MPG)'

plt.show()
```

# 3. Histograma

```python
import pandas as pd
import matplotlib.pyplot as plt

autosdf = pd.read_csv('archivos_semana_3/Automobile.csv')
autosdf = autosdf.dropna()
autosdf['mpg'].plot.hist(bins=20, color='green', edgecolor='black', figsize=(8, 6))

plt.title='Distribución del Consumo de Combustible',
plt.xlabel='Millas por Galón (MPG)',
plt.ylabel='Cantidad de Automóviles (Frecuencia)'

plt.show()
```

# 4. Boxplot

```python
#4.   Graficar un diagrama de caja (boxplot) de las características `cylinders` y `origin` para identificar posibles valores atípicos.
import pandas as pd
import matplotlib.pyplot as plt

autosdf = pd.read_csv('archivos_semana_3/Automobile.csv')
autosdf = autosdf.dropna()
autosdf.boxplot(column='cylinders', by='origin', figsize=(10, 6))

plt.title('Distribución de Cilindros por Origen')
plt.xlabel('Origen')
plt.ylabel('Cilindros')

plt.show()
```

# 5. Gráfico de barras

```python
# Realizar un gráfico de barras que muestre la distribución de automóviles según su año de modelo (`model_year`) y origen (`origin`).
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

autosdf = pd.read_csv('archivos_semana_3/Automobile.csv')
autosdf = autosdf.dropna()

sns.countplot(x='model_year', hue='origin', data=autosdf, palette='viridis')

plt.title('Distribución Según su año de modelo y origen')
plt.xlabel('Año de Modelo')
plt.ylabel('Cantidad de Vehículos')
plt.legend(title='Origen')
plt.show()

```
